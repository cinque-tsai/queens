```python
import tkinter as tk


import itertools
```

```python
class ChessBoard(tk.Frame):
    def __init__(self, parent):
        tk.Frame.__init__(self, parent)
        self.parent = parent
        self.colors = ["white", "black"]
        self.color_index = 0
        self.grid(row=0, column=0)
        self.create_board()

    def create_board(self):
        for row in range(8):
            for col in range(8):
                # create a button for each cell and set its background color
                cell = tk.Button(self, width=5, height=2, bg=self.colors[self.color_index])
                cell.grid(row=row, column=col)
                cell.bind("<Button-1>", self.change_color)
                # alternate the cell color for the next iteration
                self.color_index = 1 - self.color_index
            self.color_index = 1 - self.color_index

    def change_color(self, event):
        # get the current color of the clicked cell
        current_color = event.widget["bg"]
        # get the index of the current color in the color list
        current_color_index = self.colors.index(current_color)
        # toggle the color of the clicked cell
        new_color_index = 1 - current_color_index
        event.widget.config(bg=self.colors[new_color_index])

    def castle(self, event):
        # get the current color of the clicked cell
        current_color = event.widget["bg"]
        # get the index of the current color in the color list
        current_color_index = self.colors.index(current_color)
        # toggle the color of the clicked cell
        new_color_index = 1 - current_color_index
        event.widget.config(bg=self.colors[new_color_index])
        
    def queen(self, event):
        # get the current color of the clicked cell
        current_color = event.widget["bg"]
        # get the index of the current color in the color list
        current_color_index = self.colors.index(current_color)
        # toggle the color of the clicked cell
        new_color_index = 1 - current_color_index
        event.widget.config(bg=self.colors[new_color_index])



    
```

```python
root = tk.Tk()
root.title("Chess Board")
app = ChessBoard(root)
root.mainloop()
```

```python
class EightQueensBoard(tk.Frame):
    
    def __init__(self, master=None):
        super().__init__(master)
        
        self.master = master
        self.master.title("Eight Queens Board")
        self.grid()
        self.full = False
        self.create_board()
        self.queens = []
        
    def create_board(self):
        #row means row's index
        self.board = []
        for row in range(8):
            current_row = []
            for col in range(8):
                square = tk.Button(self, bg="white", width=7, height=3, 
                                   command=lambda row=row, col=col: self.check_queen(row, col))
                square.grid(row=row, column=col)
                current_row.append(square)
            self.board.append(current_row)
    
    #def find_queen(self,row,col):
    #    self.board[row][col]
    def check_queen(self,row,col):
        if self.full:
            error_msg = tk.Label(self, text="You can only place up to 8 queens on the board.")
            error_msg.grid(row=8, column=0, columnspan=8)
        else:
            if self.board[row][col]['bg'] =="orange":
                self.error_msg = tk.Label(self, text="A Queen has been placed here!")
                self.error_msg.grid(row=8, column=0, columnspan=8)
            else:
                try:
                    self.error_msg.grid_remove()
                except:
                    pass
                self.place_queen(row, col)
    def place_queen(self, row, col):
        if len(self.queens) < 8:
            # Place a queen on the clicked square
            
            self.queens.append((row, col))
            self.remove_possible_square(row, col)
            # Remove the possibility to place a queen on the same row, column, or diagonal
            for r in range(8):
                self.remove_possible_square(r, col)
            for c in range(8):
                self.remove_possible_square(row, c)
            if row+col <8:
                for i in range(row+col+1):
                    #print(i)
                    self.remove_possible_square(i, row+col-i)
            else:
                for i in range(row+col-7,8):
                    self.remove_possible_square(i, row+col-i)
            if row < col:
                for i in range(8-col+row):
                    self.remove_possible_square(i, i+col-row)
            else:
                for i in range(row-col,8):
                    #print(i,i+col-row)
                    self.remove_possible_square(i, i+col-row)
        
            for i,j in self.queens:
                self.board[i][j].config(text="Q", bg="orange",state="normal")
            self.full = True
            for i,j in itertools.product(range(8),range(8)):
                if self.board[i][j]['bg'] =="white":
                    #print("hi")
                    self.full = False
                    break
            if self.full:
                if len(self.queens) < 8:
                    error_msg = tk.Label(self, text=f"The board is  only filled with {len(self.queens)}QQ!")
                    error_msg.grid(row=8, column=0, columnspan=8)
                else :
                    error_msg = tk.Label(self, text="Yeah you did it!")
                    error_msg.grid(row=8, column=0, columnspan=8)

        else:
            # Display an error message if the user tries to place more than 8 queens
            error_msg = tk.Label(self, text="You can only place up to 8 queens on the board.")
            error_msg.grid(row=8, column=0, columnspan=8)
    
    #def remove_queen(self, row, col):
    #    self.board[row][col].config(state="disabled", bg="gray")

    def remove_possible_square(self, row, col):
        self.board[row][col].config(state="disabled", bg="gray")
```
