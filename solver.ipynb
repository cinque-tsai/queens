To solve 8- or n-queens problems, one must know how a chess, here queen, moves on a chessboard.
The movement of a queen is essentially a castle plus a bishop. This implies that a solution for queens must also be a solution for a castle and a bishop at the same time. Therefore, we can solve this problem by two-step: first find solutions for castles then for bishops.  


So far I geuss it's quite obvious that the castle problem is the easier one. On a n by n chessboard, the number of solution is simply n!. This discovery, though naive, can greatly reduce the computations from billions of times to tens of thousands. Furthermore, a permutation no longer needs a two-dimension list to store but one.

The Solving of n-queens problem is essentially solving n-bishops problems in a n-castles set. 

```python
class QueenSolver():
    def __init__(self):
        self.queenlist = list(itertools.permutations(range(8)))
    def solve2(self):
        qq = BishopSolver()
        #qq.solve2([6,1,5,2,0,3,7,4])
        for q in self.queenlist:
            qq.solve2(list(q))
```
