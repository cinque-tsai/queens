```python
class ChessBoard(tk.Frame):
    def __init__(self, parent):
        tk.Frame.__init__(self, parent)
        self.parent = parent
        self.colors = ["white", "black"]
        self.color_index = 0
        self.grid(row=0, column=0)
        self.create_board()

    def create_board(self):
        for row in range(8):
            for col in range(8):
                # create a button for each cell and set its background color
                cell = tk.Button(self, width=5, height=2, bg=self.colors[self.color_index])
                cell.grid(row=row, column=col)
                cell.bind("<Button-1>", self.change_color)
                # alternate the cell color for the next iteration
                self.color_index = 1 - self.color_index
            self.color_index = 1 - self.color_index

    def change_color(self, event):
        # get the current color of the clicked cell
        current_color = event.widget["bg"]
        # get the index of the current color in the color list
        current_color_index = self.colors.index(current_color)
        # toggle the color of the clicked cell
        new_color_index = 1 - current_color_index
        event.widget.config(bg=self.colors[new_color_index])

    def castle(self, event):
        # get the current color of the clicked cell
        current_color = event.widget["bg"]
        # get the index of the current color in the color list
        current_color_index = self.colors.index(current_color)
        # toggle the color of the clicked cell
        new_color_index = 1 - current_color_index
        event.widget.config(bg=self.colors[new_color_index])
        
    def queen(self, event):
        # get the current color of the clicked cell
        current_color = event.widget["bg"]
        # get the index of the current color in the color list
        current_color_index = self.colors.index(current_color)
        # toggle the color of the clicked cell
        new_color_index = 1 - current_color_index
        event.widget.config(bg=self.colors[new_color_index])



    
```
